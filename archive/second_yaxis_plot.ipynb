```python
# https://pandas.pydata.org/pandas-docs/version/0.12/visualization.html

import pandas as pd
import numpy as np

df = pd.DataFrame(np.arange(120).reshape(60,2),columns=['case','policy'])
df.iloc[:10,1] = 0
df.iloc[10:50,1] = 2
df.iloc[50:100,1] = 1
df.iloc[100:,1] = 2

ax = df.case.plot()
df.policy.plot(secondary_y=True,drawstyle='steps-post')
ax.right_ax.set_ylim(0,3)
ax.right_ax.set_yticks([])
```
